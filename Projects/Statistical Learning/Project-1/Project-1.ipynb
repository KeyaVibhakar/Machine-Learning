```python
import pandas as pd
```

#### 1. Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. (2.5 points)

```python
df = pd.read_csv("Laliga.csv",header=1, na_values=["-"] )
df = df.fillna(0)
df
```

#### 2.Print all the teams which have started playing between 1930-1980. (5 points)

```python
Debut = df['Debut'].str.extract(r'^(\d{4})', expand=False) #expand=false returns series instead of dataframe
df['DebutYear'] = pd.to_numeric(Debut)

# another way - df.loc[(df['Debut']>=1930) & (df['Debut']<=1980),'Team']
print("All the teams which have started playing between 1930-1980::\n")
df['Team'][(df['DebutYear']>=1930) & (df['DebutYear']<=1980)]
```

#### 3.Print the list of teams which came Top 5 in terms of points (2.5 points)

```python
print('List of teams which came Top 5 in terms of points :\n')
print(df.nlargest(5,'Points')['Team'])
```

#### 4.Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)

#### Goal_diff_count = GoalsFor - GoalsAgainst


```python
def Goal_diff_count(df):
        Teams_Goals = pd.DataFrame()
        Teams_Goals['Team'] = df['Team']
        Teams_Goals['GoalsFor'] = df['GoalsFor']
        Teams_Goals['GoalsAgainst'] = df['GoalsAgainst']
        Teams_Goals['Goal_diff_count'] = Teams_Goals['GoalsFor'] - Teams_Goals['GoalsAgainst']
        return Teams_Goals
    
Teams_Goals = Goal_diff_count(df)
print('Team which has maximum goal difference is :\n ', Teams_Goals.nlargest(1,'Goal_diff_count') )  
print('\n')
print('Team which has minimum goal difference is :\n ', Teams_Goals.nsmallest(1,'Goal_diff_count') )  
```

#### 5.Create a new column with name “Winning Percent” and append it to the data set (5 points)
#### Percentage of Winning = (GamesWon / GamesPlayed)*100

```python
df['Winning Percent'] = (df['GamesWon']/df['GamesPlayed'])*100
df.fillna(0, inplace=True)
df
```

#### 6. Print the top 5 teams which has the highest Winning percentage


```python
print('Top 5 teams which has highest Winning percentage are: \n')
print(df.nlargest(5,'Winning Percent')[['Team', 'Winning Percent']])
```
